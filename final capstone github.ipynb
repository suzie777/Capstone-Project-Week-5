### Explore the venues in these districts

```python
# define a function to look up the venues
def venue_info (names, latitudes, longitudes):
    radius=2000
    limit=100 

    venues_list=[]

    for name, lat, lng in zip(names, latitudes, longitudes):
        
    # create the URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
        CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, limit)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
    
        # return only relevant information for each nearby venue
        venues_list.append([(name,  
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Zip code', 'Venue', 'Venue Latitude', 'Venue Longitude', 'Venue Category']
    
    return(nearby_venues)
```

```python
# explore the venues in the selected districts
buffalo_venues = venue_info(names=df['Zip code'],
                                latitudes=df['Latitude'],
                                longitudes=df['Longitude']
                                  )
```

```python
buffalo_onehot = pd.get_dummies(buffalo_venues[['Venue Category']], prefix="", prefix_sep="")

# add the column back to dataframe
buffalo_onehot['Zip code'] = buffalo_venues['Zip code'] 

# move the column to the first column
fixed_columns = [buffalo_onehot.columns[-1]] + list(buffalo_onehot.columns[:-1])
buffalo_onehot = buffalo_onehot[fixed_columns]
```

```python
buffalo_venues_grouped = buffalo_onehot.groupby('Zip code').mean().reset_index()
buffalo_venues_grouped
```

```python
# What are the top 5 venues in each district
num = 5

for code in buffalo_venues_grouped['Zip code']:
    print('Zip code:', code)
    temp = buffalo_venues_grouped[buffalo_venues_grouped['Zip code'] == code].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num))
    print('\n')
```

```python
# define a function to get the most popular venues
def popular_venues(row, num):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num]
```

```python
num=10
indicators = ['st', 'nd', 'rd']

# create columns according to number of most popular venues
columns = ['Zip code']
for ind in np.arange(num):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
venues_sorted = pd.DataFrame(columns=columns)
venues_sorted['Zip code'] = buffalo_venues_grouped['Zip code']

for ind in np.arange(buffalo_venues_grouped.shape[0]):
    venues_sorted.iloc[ind, 1:] = popular_venues(buffalo_venues_grouped.iloc[ind, :], num)

venues_sorted
```

### K-means clustering

```python
# Import kmeans
from sklearn.cluster import KMeans
```

```python
kclusters = 3

buffalo_grouped_clustering = buffalo_venues_grouped.drop('Zip code', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(buffalo_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
venues_sorted=venues_sorted.insert(0, 'Cluster Label', kmeans.labels_)
```

```python
# merge buffalo_grouped with buffalo data to add latitude and longitude
buffalo_venues_grouped_clustering = df.join(venues_sorted.set_index('Zip code'), on='Zip code')
buffalo_venues_grouped_clustering.head() 
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# create map
map_clusters = folium.Map(location=[42.8864, -78.8784], zoom_start=14)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(buffalo_venues_grouped_clustering['Latitude'], 
                                  buffalo_venues_grouped_clustering['Longitude'], 
                                  buffalo_venues_grouped_clustering['Zip code'], 
                                  buffalo_venues_grouped_clustering['Cluster Label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=8,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
